```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.columns
```

```python
# Barplot with summary statistics: Average fare by class
sns.barplot(x='Pclass', y='Fare', data=df)

plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()
```

```python
# Simple Barplot: Count of passengers by class
sns.countplot(x='Pclass', data=df)

plt.title('Count of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Barplot with hue: Survival count by class and gender
sns.countplot(x='Pclass', hue='Sex', data=df)

plt.title('Count of Passengers by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
# Create a box plot to visualize the fare distribution across different passenger classes
sns.boxplot(x='Pclass', y='Fare', data=df)

plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Create a box plot to visualize the fare distribution across different passenger classes
sns.boxplot(x='Pclass', y='Fare',hue='Sex', data=df)

plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Create a KDE plot for the 'Age' variable
sns.kdeplot(df['Age'].dropna(), shade=True)

# Add title and labels
plt.title('Age Distribution - KDE Plot')
plt.xlabel('Age')
plt.ylabel('Density')
```

```python
# Create KDE plots for age distribution of survivors and non-survivors
sns.kdeplot(df[df['Survived'] == 1]['Age'].dropna(), shade=True, label='Survived')
sns.kdeplot(df[df['Survived'] == 0]['Age'].dropna(), shade=True, label='Did Not Survive')

# Add title and labels
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')

# Show legend and plot
plt.legend()
plt.show()
```

```python
# Create a scatter plot of Age vs. Fare
sns.scatterplot(x='Age', y='Fare', data=df)

plt.title('Scatter Plot of Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Scatter plot of Age vs. Fare with 'Survived' as the category
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)

plt.title('Scatter Plot of Age vs. Fare by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()
```

```python
# Count the number of passengers in each class
class_counts = df['Pclass'].value_counts()

# Create a pie chart
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)

# Add title
plt.title('Passenger Class Distribution')

# Display the pie chart
plt.show()
```

```python
# Calculate correlation matrix for numerical features
corr_matrix = df.select_dtypes(include=['int64', 'float64']).corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add title
plt.title('Heatmap of Correlation Matrix')

# Show the plot
plt.show()
```
